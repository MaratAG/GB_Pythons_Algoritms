РЕЗУЛЬТАТЫ ВЫПОЛНЕНИЯ ТЕСТА (А):

```python
from matplotlib import pyplot as plt
```

ДАННЫЕ ПО ОЦЕНКЕ СКОРОСТИ ВЫПОЛНЕНИЯ АЛГОРИТМОВ

```python
size_of_array = [1000, 10000, 100000, 500000, 750000, 1000000]

time_of_test_1 = [.00169, .0139, .146, .740, 1.11, 1.49]   
time_of_test_2 = [.00144, .014, .143, .768, 1.2, 1.53]  
time_of_test_3 = [.00136, 0.0123, 0.127, .644, .991, 1.33] 
```

```python
plt.plot(size_of_array, time_of_test_1, color='green', marker='o', linestyle='solid', label='1')
plt.plot(size_of_array, time_of_test_2, color='red', marker='o', linestyle='solid', label='2')
plt.plot(size_of_array, time_of_test_3, color='blue', marker='o', linestyle='solid', label='3')

plt.title('Оценка скорости выполнения алгоритмов (timeit)')
plt.ylabel('Время (секунды)')
plt.xlabel('Размер масcива (элементы)')

plt.legend(loc=0)

plt.show()
```
