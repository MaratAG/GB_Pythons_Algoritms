РЕЗУЛЬТАТЫ ВЫПОЛНЕНИЯ ТЕСТА (А):

```python
from matplotlib import pyplot as plt
```

ДАННЫЕ ПО ОЦЕНКЕ СКОРОСТИ ВЫПОЛНЕНИЯ АЛГОРИТМОВ

```python
size_of_array = [100, 500, 1000, 1500, 2000]

time_of_test_1 = [.0039, .237, 1.13, 2.93, 5.76]   
time_of_test_2 = [.0038, .04, .0473, .515, .568]  
```

```python
plt.plot(size_of_array, time_of_test_1, color='green', marker='o', linestyle='solid', label='Classic')
plt.plot(size_of_array, time_of_test_2, color='red', marker='o', linestyle='solid', label='Sieve')

plt.title('Оценка скорости выполнения алгоритмов (timeit)')
plt.ylabel('Время (секунды)')
plt.xlabel('Номер позиции искомого простого числа')

plt.legend(loc=0)

plt.show()
```

2. РЕЗУЛЬТАТЫ ВЫПОЛНЕНИЯ ТЕСТА (Б):

```python
size_of_array = [100, 500, 1000, 1500, 2000]

part_in_time_1 = [.012, .313, 1.334, 3.421, 6.655]   
part_in_time_2 = [.008, .057, .063, .527, .594]  
```

```python
plt.plot(size_of_array, part_in_time_1, color='green', marker='o', linestyle='solid', label='Prime')
plt.plot(size_of_array, part_in_time_2, color='red', marker='o', linestyle='solid', label='Sieve')

plt.title('Оценка производительности выполнения алгоритмов (cProfile)')
plt.ylabel('Доля в общем времени выполнения программы')
plt.xlabel('Номер позиции искомого простого числа')

plt.legend(loc=0)

plt.show()
```
